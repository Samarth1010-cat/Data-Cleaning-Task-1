```python
# ==========================================================
# TASK 1 - DATA CLEANING AND PREPROCESSING
# Data Analyst Internship
# ==========================================================

# ----------------------------
# Import Libraries
# ----------------------------

import pandas as pd
import numpy as np

print("="*60)
print("TASK 1 - DATA CLEANING AND PREPROCESSING")
print("="*60)

# ----------------------------
# Load Dataset
# ----------------------------

# Dataset is TAB separated
df = pd.read_csv("customer_data.csv", sep="\t")

print("\nDataset Loaded Successfully!\n")

# ==========================================================
# BEFORE CLEANING
# ==========================================================

print("="*60)
print("FIRST 5 ROWS")
print("="*60)
display(df.head())

print("="*60)
print("DATASET INFORMATION")
print("="*60)
df.info()

print("\n")

print("="*60)
print("DATASET SHAPE")
print("="*60)
print(df.shape)

print("\n")

print("="*60)
print("COLUMN NAMES")
print("="*60)
print(df.columns.tolist())

print("\n")

print("="*60)
print("MISSING VALUES")
print("="*60)
print(df.isnull().sum())

print("\n")

print("="*60)
print("DUPLICATE ROWS")
print("="*60)
print(df.duplicated().sum())

# ==========================================================
# DATA CLEANING
# ==========================================================

print("\n")
print("="*60)
print("STARTING DATA CLEANING...")
print("="*60)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing Income values with median
if "Income" in df.columns:
    df["Income"] = df["Income"].fillna(df["Income"].median())

# Clean text columns
text_columns = ["Education", "Marital_Status"]

for col in text_columns:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.title()
        )

# Rename columns
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
)

# Convert date column
if "dt_customer" in df.columns:
    df["dt_customer"] = pd.to_datetime(
        df["dt_customer"],
        format="%d-%m-%Y",
        errors="coerce"
    )

# ==========================================================
# AFTER CLEANING
# ==========================================================

print("\n")
print("="*60)
print("AFTER CLEANING")
print("="*60)

print("\nDataset Shape:")
print(df.shape)

print("\n")

print("Missing Values Remaining:")
print(df.isnull().sum())

print("\n")

print("Duplicate Rows Remaining:")
print(df.duplicated().sum())

print("\n")

print("Updated Column Names:")
print(df.columns.tolist())

print("\n")

print("Updated Data Types:")
print(df.dtypes)

print("\n")

print("First 5 Rows After Cleaning:")
display(df.head())

# ==========================================================
# SAVE CLEANED DATASET
# ==========================================================

df.to_csv("cleaned_customer_data.csv", index=False)

print("\n")
print("="*60)
print("PROJECT COMPLETED SUCCESSFULLY")
print("="*60)

print("\nCleaned dataset saved as:")
print("cleaned_customer_data.csv")
```
